```
df_train_term = pd.read_csv('/content/gdrive/My Drive/impact/train_term.csv')
df_val_term = pd.read_csv('/content/gdrive/My Drive/impact/val_term.csv')
df_test_term = pd.read_csv('/content/gdrive/My Drive/impact/test_term.csv')

df_train_vl1 = pd.read_csv('/content/gdrive/My Drive/impact/train_vl_1.csv')
df_val_vl1 = pd.read_csv('/content/gdrive/My Drive/impact/val_vl_1.csv')
df_test_vl1 = pd.read_csv('/content/gdrive/My Drive/impact/test_vl_1.csv')

df_train_vl2 = pd.read_csv('/content/gdrive/My Drive/impact/train_vl_2.csv')
df_val_vl2 = pd.read_csv('/content/gdrive/My Drive/impact/val_vl_2.csv')
df_test_vl2 = pd.read_csv('/content/gdrive/My Drive/impact/test_vl_2.csv')

df_train_vl3 = pd.read_csv('/content/gdrive/My Drive/impact/train_vl_3.csv')
df_val_vl3 = pd.read_csv('/content/gdrive/My Drive/impact/val_vl_3.csv')
df_test_vl3 = pd.read_csv('/content/gdrive/My Drive/impact/test_vl_3.csv')

df_train_most_depth = pd.read_csv('/content/gdrive/My Drive/impact/train_most_depth.csv')
df_val_most_depth = pd.read_csv('/content/gdrive/My Drive/impact/val_most_depth.csv')
df_test_most_depth = pd.read_csv('/content/gdrive/My Drive/impact/test_most_depth.csv')
```

```
df_train_vl1.drop(columns=['id'], inplace=True)
df_train_vl2.drop(columns=['id'], inplace=True)
df_train_vl3.drop(columns=['id'], inplace=True)

df_val_vl1.drop(columns=['id'], inplace=True)
df_val_vl2.drop(columns=['id'], inplace=True)
df_val_vl3.drop(columns=['id'], inplace=True)

df_test_vl1.drop(columns=['id'], inplace=True)
df_test_vl2.drop(columns=['id'], inplace=True)
df_test_vl3.drop(columns=['id'], inplace=True)
```

Prepare train, val and test data containing all unique columns of "Term", "Variable Length 1", "Variable Length 2" and "Variable Length 3".

Train

```
# Merge columns by abstract
df3 = df_train_term.merge(df_train_vl1,how='outer',on='abstract',suffixes=['','_x'])
overlapping_sufx = df3.columns.values[df3.columns.str.endswith('_x')]
for sufx in overlapping_sufx:
    m = sufx[:-2]
    df3[m][df3[sufx].notnull()] = df3[m].fillna(0) + df3[sufx]
    df3=df3.drop(columns=sufx)

df4 = df3.merge(df_train_vl2,how='outer',on='abstract',suffixes=['','_x'])
overlapping_sufx = df4.columns.values[df4.columns.str.endswith('_x')]
for sufx in overlapping_sufx:
    m = sufx[:-2]
    df4[m][df4[sufx].notnull()] = df4[m].fillna(0) + df4[sufx]
    df4=df4.drop(columns=sufx)

df5 = df4.merge(df_train_vl3,how='outer',on='abstract',suffixes=['','_x'])
overlapping_sufx = df5.columns.values[df5.columns.str.endswith('_x')]
for sufx in overlapping_sufx:
    m = sufx[:-2]
    df5[m][df5[sufx].notnull()] = df5[m].fillna(0) + df5[sufx]
    df5=df5.drop(columns=sufx)
df5.head()
```

```
df5 = df5.fillna(0)
cols = df5.columns
for i in cols:
  if df5[i].dtype == 'float64':
    df5[i] = df5[i].astype('int64')
```
